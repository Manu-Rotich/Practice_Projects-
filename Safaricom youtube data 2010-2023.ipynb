3. Explore the distribution of different variables

```python
sns.histplot(df['Views'], bins=20)
plt.show()
```

```python
def calculate_engagement_rate(Likes, dislikes):
  if dislikes == 0:
    return Likes  # Avoid division by zero
  else:
    return (Likes / (Likes + dislikes)) * 100
  print(calculate_engagement_rate(Likes, dislikes))  
```

```python
sns.histplot(df["Views"], bins=20, kde=True)
plt.title("Distribution of Video Views")
plt.xlabel("Number of Views")
plt.ylabel("Density")
plt.show()
```

4. Analyze the relationship between variables

Scatter Plots:

Create scatter plots to visually explore the relationships between numerical variables. Use color or size to represent categorical variables (e.g., title length).

```python
sns.scatterplot(x='Likes', y='Comments', data=df)
plt.show()
sns.scatterplot(x='Published_Year', y='Views', data=df)
plt.show()

```

Converts "Published Date" to DateTime format.

It first uses pd.to_datetime(df['Published Date']) to convert the values in the "Published Date" column to DateTime objects.
This assumes that the "Published Date" column contains strings in a format that can be parsed as dates or times (e.g., "2023-10-26", "2023-10-26 15:30:00").
If the format is different, you might need to adjust the parsing logic
Then, it uses .dt.year to access the year component of each DateTime object. 
This results in a new Series containing only the years, which is then assigned to the newly created column "Published_Year"
In essence, this code snippet adds a new column to your DataFrame that stores the year extracted from the original "Published Date" column.
This can be useful for further analysis or visualizations based on the year of publication.

```python
df['Published_Year'] = pd.to_datetime(df['Published Date']).dt.year
```

A heatmap is a powerful visual tool that uses color intensity to represent the magnitude of values within a matrix or table. 
It helps you quickly grasp trends, patterns, and relationships between variables, especially when dealing with large datasets.

```python
heatmap_data = df.groupby('Published_Year')['Views'].sum()
```

```python
heatmap_data = df.pivot_table(index='Published_Year', values='Views', aggfunc='sum')
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt='.2f')
plt.xlabel('Published Year')
plt.ylabel('Total Views')  # Adjust label if needed
plt.title('Distribution of Total Video Views by Published Year')
plt.xticks(rotation=45)
plt.show()
```
